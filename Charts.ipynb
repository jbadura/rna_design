```python
from matplotlib import pyplot as plt
import seaborn as sns
import os
import pandas as pd
plt.rcParams['figure.figsize'] = [14, 14]

data = pd.read_csv("results/plots_data.csv", sep=';')

def draw_plots(df):
    fig, ax = plt.subplots(3,1)
    sns.violinplot(data=df, y="rnapdist", x="algo", ax=ax[0])
    sns.violinplot(data=df, y="seqidentity", x="algo", ax=ax[1])
    sns.violinplot(data=df, y="rnadistance", x="algo", ax=ax[2])
```

```python
filtered = data[data['is_extended'] == 0]
draw_plots(filtered)
```

```python
filtered = data[data['is_extended'] == 1]
draw_plots(filtered)
```

```python
filtered = data[data['is_extended'] == 0]
filtered = filtered[filtered['type'] == 'Internal loop']
draw_plots(filtered)
```

```python
filtered = data[data['is_extended'] == 1]
filtered = filtered[filtered['type'] == 'Internal loop']
draw_plots(filtered)
```
